9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
yangon = df[df['City'] == 'Yangon']
mandalay = df[df['City'] == 'Mandalay']

total_yangon = len(yangon)
total_mandalay = len(mandalay)

count_yangon = yangon['Product line'].value_counts()
count_mandalay = mandalay['Product line'].value_counts()

yangon_percent = count_yangon / total_yangon
mandalay_percent = count_mandalay / total_mandalay

print('percentage distribution of the number of invoices by product categories in Yangon:\n', '\n', yangon_percent)
print()
print('percentage distribution of the number of invoices by product categories in Mandalay:\n', '\n', mandalay_percent)
```
